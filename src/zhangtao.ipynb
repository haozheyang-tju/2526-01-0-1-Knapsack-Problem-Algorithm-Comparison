````python


"""
0/1 èƒŒåŒ…å®éªŒåŸºå‡†è„šæœ¬ - æ¨¡æ‹Ÿé€€ç« vs è®°å¿†æ¨¡æ‹Ÿé€€ç«å¯¹æ¯”å®éªŒ

ç”¨é€”ï¼š
- æä¾›ä¸åŸå§‹å®éªŒç›¸åŒçš„å®ä¾‹é›†ç”Ÿæˆæ–¹å¼
- å®ç°æ ‡å‡†æ¨¡æ‹Ÿé€€ç«ç®—æ³•(SA)å’Œè®°å¿†æ¨¡æ‹Ÿé€€ç«ç®—æ³•(Memetic SA)
- è¿›è¡Œè¯¦ç»†çš„æ€§èƒ½å¯¹æ¯”åˆ†æ

æœ¬å®éªŒä¿æŒä¸åŸå§‹è´ªå¿ƒç®—æ³•å®éªŒå®Œå…¨ç›¸åŒçš„å®ä¾‹é›†ï¼Œä»…æ›¿æ¢å¯¹æ¯”ç®—æ³•
"""

import math
import random
import time
import os
import json
from dataclasses import dataclass
from typing import List, Tuple, Dict, Callable, Optional
from collections import defaultdict

import numpy as np
import pandas as pd
import seaborn as sns
import zlib


import matplotlib.pyplot as plt
from datetime import datetime
from scipy import stats
from matplotlib import rcParams



#  ä¸€ã€ç¨³å®šå“ˆå¸Œ & å…¨å±€è®¾ç½® - ä¸åŸå§‹å®éªŒå®Œå…¨ç›¸åŒ


def stable_hash_str(*parts) -> int:
    """è·¨å¹³å°/è·¨è¿›ç¨‹ç¨³å®šçš„ 32 ä½å“ˆå¸Œï¼Œç”¨äºç”Ÿæˆå¯å¤ç°çš„éšæœºç§å­ã€‚"""
    s = "|".join(str(p) for p in parts)
    return zlib.adler32(s.encode("utf-8")) & 0xffffffff

# å›ºå®šé¡¶å±‚éšæœºç§å­ï¼ˆåº“çº§ï¼‰
random.seed(42)
np.random.seed(42)

OUTDIR = "sa_memetic_comparison"
os.makedirs(OUTDIR, exist_ok=True)


#  äºŒã€æ•°æ®ç»“æ„å®šä¹‰ - ä¸åŸå§‹å®éªŒå®Œå…¨ç›¸åŒ


@dataclass
class Item:
    """ç‰©å“ç»“æ„ï¼šw=é‡é‡, v=ä»·å€¼, idx=åŸå§‹ç´¢å¼•ï¼ˆç”¨äºå›æº¯/è°ƒè¯•ï¼‰"""
    w: int   # weight
    v: int   # value
    idx: int # åŸå§‹ç´¢å¼•

# æ‰€æœ‰ç®—æ³•ç»Ÿä¸€æ¥å£ï¼š (items, C) -> (value, chosen_indices)
AlgoFn = Callable[[List[Item], int], Tuple[int, List[int]]]


#  ä¸‰ã€ç®—æ³•å®ç° - æ¨¡æ‹Ÿé€€ç«(SA)å’Œè®°å¿†æ¨¡æ‹Ÿé€€ç«(Memetic SA)


def simulated_annealing_knapsack_fast(items: List[Item], C: int) -> Tuple[int, List[int]]:
    """
    å¿«é€Ÿæ¨¡æ‹Ÿé€€ç«ç®—æ³• - ä¼˜åŒ–å‚æ•°ä»¥å‡å°‘è¿è¡Œæ—¶é—´
    
    ä¸»è¦ä¼˜åŒ–ï¼š
    1. é™ä½è¿­ä»£æ¬¡æ•°
    2. æé«˜å†·å´é€Ÿç‡
    3. å‡å°‘æœ€å¤§åœæ»æ¬¡æ•°
    """
    n = len(items)
    
    # ä¼˜åŒ–åçš„å‚æ•°è®¾ç½® - å¤§å¹…å‡å°‘è¿è¡Œæ—¶é—´
    if n <= 50:
        initial_temp = 300      # é™ä½åˆå§‹æ¸©åº¦
        cooling_rate = 0.97     # æé«˜å†·å´é€Ÿç‡ï¼Œæ›´å¿«é™æ¸©
        iterations_per_temp = 30  # å¤§å¹…å‡å°‘æ¯ä¸ªæ¸©åº¦çš„è¿­ä»£æ¬¡æ•°
        max_stagnation = 10     # å‡å°‘æœ€å¤§åœæ»æ¬¡æ•°
    elif n <= 100:
        initial_temp = 200
        cooling_rate = 0.96
        iterations_per_temp = 20
        max_stagnation = 8
    else:  # å¤§è§„æ¨¡å®ä¾‹
        initial_temp = 150
        cooling_rate = 0.95
        iterations_per_temp = 15  # è¿›ä¸€æ­¥å‡å°‘
        max_stagnation = 5
    
    # åˆå§‹åŒ–å½“å‰è§£ - ç®€åŒ–ç‰ˆè´ªå¿ƒ
    def initialize_solution():
        density_sorted = sorted(range(n), key=lambda i: items[i].v / items[i].w, reverse=True)
        solution = [0] * n
        current_weight = 0
        
        # åªé€‰æ‹©å‰å‡ ä¸ªç‰©å“ï¼Œå‡å°‘åˆå§‹åŒ–æ—¶é—´
        max_items = min(20, n)  # é™åˆ¶åˆå§‹åŒ–ç‰©å“æ•°é‡
        for idx in density_sorted[:max_items]:
            if current_weight + items[idx].w <= C:
                solution[idx] = 1
                current_weight += items[idx].w
        return solution
    
    def evaluate(solution):
        total_value = 0
        total_weight = 0
        for i in range(n):
            if solution[i] == 1:
                total_value += items[i].v
                total_weight += items[i].w
        if total_weight > C:
            return max(0, total_value - 1000 * (total_weight - C))
        return total_value
    
    # ç®€åŒ–é‚»åŸŸæ“ä½œ
    def generate_neighbor_fast(current_solution):
        new_solution = current_solution.copy()
        
        # åªä½¿ç”¨æœ€ç®€å•çš„ç¿»è½¬æ“ä½œï¼Œå‡å°‘è®¡ç®—
        if random.random() < 0.7:  # 70%æ¦‚ç‡ä½¿ç”¨ç®€å•ç¿»è½¬
            idx = random.randint(0, n-1)
            new_solution[idx] = 1 - new_solution[idx]
        else:  # 30%æ¦‚ç‡ä½¿ç”¨swapæ“ä½œ
            idx1, idx2 = random.sample(range(n), 2)
            new_solution[idx1], new_solution[idx2] = new_solution[idx2], new_solution[idx1]
        
        return new_solution
    
    def repair_solution_fast(solution):
        current_weight = sum(items[i].w for i in range(n) if solution[i] == 1)
        
        # å¦‚æœè¶…é‡ï¼Œç§»é™¤æœ€å°‘ç‰©å“è¾¾åˆ°å¯è¡Œ
        if current_weight > C:
            selected_indices = [i for i in range(n) if solution[i] == 1]
            if selected_indices:
                # æŒ‰ä»·å€¼å¯†åº¦å‡åºæ’åºï¼Œç§»é™¤ä½å¯†åº¦ç‰©å“
                sorted_by_density = sorted(
                    selected_indices,
                    key=lambda i: items[i].v / items[i].w
                )
                # åªç§»é™¤å¿…è¦çš„æœ€å°‘ç‰©å“
                for idx in sorted_by_density:
                    if current_weight <= C:
                        break
                    solution[idx] = 0
                    current_weight -= items[idx].w
        return solution
    
    # æ¨¡æ‹Ÿé€€ç«ä¸»å¾ªç¯ - ä¼˜åŒ–ç‰ˆ
    current_solution = initialize_solution()
    current_value = evaluate(current_solution)
    best_solution = current_solution.copy()
    best_value = current_value
    
    temperature = initial_temp
    stagnation_count = 0
    max_iterations = 1000  # æ·»åŠ æ€»è¿­ä»£æ¬¡æ•°é™åˆ¶
    
    iteration = 0
    while temperature > 1 and stagnation_count < max_stagnation and iteration < max_iterations:
        improved = False
        
        for _ in range(iterations_per_temp):
            # ç”Ÿæˆæ–°è§£
            new_solution = generate_neighbor_fast(current_solution)
            new_solution = repair_solution_fast(new_solution)
            new_value = evaluate(new_solution)
            
            delta = current_value - new_value
            
            # Metropoliså‡†åˆ™
            if delta < 0 or random.random() < math.exp(-delta / temperature):
                current_solution = new_solution
                current_value = new_value
                
                if new_value > best_value:
                    best_solution = new_solution.copy()
                    best_value = new_value
                    improved = True
                    stagnation_count = 0
            
            iteration += 1
        
        if not improved:
            stagnation_count += 1
            
        temperature *= cooling_rate
        
        # æå‰é€€å‡ºæ¡ä»¶
        if iteration >= max_iterations:
            break
    
    # å¿«é€Ÿå±€éƒ¨ä¼˜åŒ–
    def quick_local_optimize(solution):
        """å¯¹è§£è¿›è¡Œå¿«é€Ÿå±€éƒ¨ä¼˜åŒ–"""
        improved = True
        while improved:
            improved = False
            # å°è¯•æ·»åŠ ä¸€ä¸ªæœªé€‰ä¸­çš„é«˜ä»·å€¼å¯†åº¦ç‰©å“
            unselected = [i for i in range(n) if solution[i] == 0]
            selected = [i for i in range(n) if solution[i] == 1]
            current_weight = sum(items[i].w for i in selected)
            
            # æŒ‰ä»·å€¼å¯†åº¦æ’åº
            unselected_sorted = sorted(unselected, key=lambda i: items[i].v / items[i].w, reverse=True)
            
            for idx in unselected_sorted[:10]:  # åªæ£€æŸ¥å‰10ä¸ª
                if current_weight + items[idx].w <= C:
                    # æ£€æŸ¥æ˜¯å¦å¯ä»¥æ·»åŠ 
                    solution[idx] = 1
                    current_weight += items[idx].w
                    improved = True
                    break
        
        return solution
    
    # å¯¹æœ€ç»ˆè§£è¿›è¡Œå¿«é€Ÿå±€éƒ¨ä¼˜åŒ–
    best_solution = quick_local_optimize(best_solution)
    best_value = evaluate(best_solution)
    
    chosen_indices = [items[i].idx for i in range(n) if best_solution[i] == 1]
    return best_value, chosen_indices

def memetic_simulated_annealing_fast(items: List[Item], C: int) -> Tuple[int, List[int]]:
    """
    å¿«é€Ÿè®°å¿†æ¨¡æ‹Ÿé€€ç«ç®—æ³• - å¤§å¹…ä¼˜åŒ–è¿è¡Œæ—¶é—´
    
    ä¸»è¦ä¼˜åŒ–ï¼š
    1. å‡å°ç§ç¾¤è§„æ¨¡
    2. å‡å°‘å±€éƒ¨æœç´¢è¿­ä»£æ¬¡æ•°
    3. ç®€åŒ–äº¤å‰å’Œå˜å¼‚æ“ä½œ
    """
    n = len(items)
    
    # ä¼˜åŒ–åçš„å‚æ•°è®¾ç½®
    if n <= 50:
        initial_temp = 300
        cooling_rate = 0.97
        iterations_per_temp = 30
        max_stagnation = 10
        population_size = 5      # å‡å°ç§ç¾¤è§„æ¨¡
        local_search_iterations = 5  # å‡å°‘å±€éƒ¨æœç´¢è¿­ä»£æ¬¡æ•°
    elif n <= 100:
        initial_temp = 200
        cooling_rate = 0.96
        iterations_per_temp = 20
        max_stagnation = 8
        population_size = 4
        local_search_iterations = 4
    else:
        initial_temp = 150
        cooling_rate = 0.95
        iterations_per_temp = 15
        max_stagnation = 5
        population_size = 3
        local_search_iterations = 3
    
    # å¿«é€Ÿåˆå§‹åŒ–ç§ç¾¤
    def initialize_population_fast():
        population = []
        
        # ç­–ç•¥1: è´ªå¿ƒå¯†åº¦ç­–ç•¥ï¼ˆå¿«é€Ÿç‰ˆï¼‰
        density_sorted = sorted(range(n), key=lambda i: items[i].v / items[i].w, reverse=True)
        solution = [0] * n
        current_weight = 0
        max_items = min(15, n)  # é™åˆ¶åˆå§‹åŒ–ç‰©å“æ•°é‡
        for idx in density_sorted[:max_items]:
            if current_weight + items[idx].w <= C:
                solution[idx] = 1
                current_weight += items[idx].w
        population.append(solution)
        
        # ç­–ç•¥2: éšæœºç­–ç•¥ï¼ˆå°‘é‡ï¼‰
        for _ in range(population_size - 1):
            solution = [random.randint(0, 1) for _ in range(n)]
            solution = repair_solution_fast(solution)
            population.append(solution)
        
        return population
    
    def evaluate(solution):
        total_value = 0
        total_weight = 0
        for i in range(n):
            if solution[i] == 1:
                total_value += items[i].v
                total_weight += items[i].w
        if total_weight > C:
            return max(0, total_value - 1000 * (total_weight - C))
        return total_value
    
    def repair_solution_fast(solution):
        solution = solution.copy()
        current_weight = sum(items[i].w for i in range(n) if solution[i] == 1)
        
        if current_weight > C:
            selected_indices = [i for i in range(n) if solution[i] == 1]
            if selected_indices:
                sorted_by_density = sorted(
                    selected_indices,
                    key=lambda i: items[i].v / items[i].w
                )
                for idx in sorted_by_density:
                    if current_weight <= C:
                        break
                    solution[idx] = 0
                    current_weight -= items[idx].w
        return solution
    
    # å¿«é€Ÿå±€éƒ¨æœç´¢
    def local_search_fast(solution, max_iterations):
        best_solution = solution.copy()
        best_value = evaluate(best_solution)
        
        # å‡å°‘å±€éƒ¨æœç´¢å°è¯•æ¬¡æ•°
        for _ in range(min(max_iterations, 5)):  # æœ€å¤š5æ¬¡å°è¯•
            neighbor = best_solution.copy()
            idx = random.randint(0, n-1)
            neighbor[idx] = 1 - neighbor[idx]
            neighbor = repair_solution_fast(neighbor)
            neighbor_value = evaluate(neighbor)
            
            if neighbor_value > best_value:
                best_solution = neighbor
                best_value = neighbor_value
        
        return best_solution
    
    # ç®€åŒ–äº¤å‰æ“ä½œ
    def crossover_fast(parent1, parent2):
        if n <= 1:
            return parent1.copy(), parent2.copy()
        
        # å‡åŒ€äº¤å‰
        child1 = [0] * n
        child2 = [0] * n
        
        for i in range(n):
            if random.random() < 0.5:
                child1[i] = parent1[i]
                child2[i] = parent2[i]
            else:
                child1[i] = parent2[i]
                child2[i] = parent1[i]
        
        return child1, child2
    
    # ç®€åŒ–å˜å¼‚æ“ä½œ
    def mutate_fast(solution, mutation_rate=0.05):  # é™ä½å˜å¼‚ç‡
        mutated = solution.copy()
        
        # åªå˜å¼‚å°‘é‡ä½ç½®
        mutation_count = max(1, int(n * mutation_rate))
        for _ in range(mutation_count):
            idx = random.randint(0, n-1)
            mutated[idx] = 1 - mutated[idx]
        
        mutated = repair_solution_fast(mutated)
        return mutated
    
    # é”¦æ ‡èµ›é€‰æ‹©ï¼ˆç®€åŒ–ç‰ˆï¼‰
    def tournament_selection_fast(population, tournament_size=2):  # å‡å°é”¦æ ‡èµ›è§„æ¨¡
        tournament_indices = random.sample(range(len(population)), tournament_size)
        tournament_solutions = [population[i] for i in tournament_indices]
        tournament_values = [evaluate(s) for s in tournament_solutions]
        
        best_idx = tournament_indices[tournament_values.index(max(tournament_values))]
        return population[best_idx]
    
    # ä¸»ç®—æ³•æµç¨‹ï¼ˆä¼˜åŒ–ç‰ˆï¼‰
    population = initialize_population_fast()
    population_values = [evaluate(s) for s in population]
    best_solution = population[np.argmax(population_values)].copy()
    best_value = max(population_values)
    
    temperature = initial_temp
    stagnation_count = 0
    max_generations = 50  # é™åˆ¶æœ€å¤§ä»£æ•°
    
    generation = 0
    while temperature > 1 and stagnation_count < max_stagnation and generation < max_generations:
        new_population = []
        
        # ä¿ç•™ç²¾è‹±
        elite_idx = np.argmax(population_values)
        new_population.append(population[elite_idx].copy())
        
        improved = False
        
        # ç”Ÿæˆæ–°ä¸€ä»£
        while len(new_population) < population_size:
            parent1 = tournament_selection_fast(population)
            parent2 = tournament_selection_fast(population)
            
            child1, child2 = crossover_fast(parent1, parent2)
            
            child1 = mutate_fast(child1)
            child2 = mutate_fast(child2)
            
            # å¿«é€Ÿå±€éƒ¨æœç´¢
            child1 = local_search_fast(child1, local_search_iterations)
            child2 = local_search_fast(child2, local_search_iterations)
            
            # è¯„ä¼°å¹¶æ¥å—
            for child in [child1, child2]:
                child_value = evaluate(child)
                parent_value = evaluate(parent1)
                
                delta = parent_value - child_value
                
                if delta < 0 or random.random() < math.exp(-delta / temperature):
                    new_population.append(child)
                    if child_value > best_value:
                        best_solution = child.copy()
                        best_value = child_value
                        improved = True
                        stagnation_count = 0
        
        population = new_population[:population_size]
        population_values = [evaluate(s) for s in population]
        
        if not improved:
            stagnation_count += 1
            
        temperature *= cooling_rate
        generation += 1
    
    # æœ€ç»ˆå¿«é€Ÿä¼˜åŒ–
    def final_quick_optimize(solution):
        """æœ€ç»ˆå¿«é€Ÿä¼˜åŒ–"""
        current_weight = sum(items[i].w for i in range(n) if solution[i] == 1)
        
        # å°è¯•æ·»åŠ æœªé€‰ä¸­çš„é«˜ä»·å€¼ç‰©å“
        unselected = [i for i in range(n) if solution[i] == 0]
        # æŒ‰ä»·å€¼å¯†åº¦æ’åº
        unselected_sorted = sorted(unselected, key=lambda i: items[i].v / items[i].w, reverse=True)
        
        for idx in unselected_sorted[:15]:  # åªæ£€æŸ¥å‰15ä¸ª
            if current_weight + items[idx].w <= C:
                solution[idx] = 1
                current_weight += items[idx].w
        
        return solution
    
    best_solution = final_quick_optimize(best_solution)
    best_value = evaluate(best_solution)
    
    chosen_indices = [items[i].idx for i in range(n) if best_solution[i] == 1]
    return best_value, chosen_indices

# === æ³¨å†Œè¦è¯„æµ‹çš„ç®—æ³• ===
ALGOS: Dict[str, AlgoFn] = {
    "SA_Fast": simulated_annealing_knapsack_fast,
    "MemeticSA_Fast": memetic_simulated_annealing_fast,
}


#  å››ã€åŸºå‡†ç®—æ³• - ä¸åŸå§‹å®éªŒå®Œå…¨ç›¸åŒ


def dp_optimal(items: List[Item], C: int) -> Tuple[int, List[int]]:
    """å®¹é‡ç»´åº¦ DPï¼ŒO(nC)ã€‚è§„æ¨¡è¿‡å¤§æ—¶åªè¿”å›å€¼ï¼Œä¸å›æº¯ã€‚"""
    n = len(items)
    dp = [0] * (C + 1)

    # å¯ç”¨å›æº¯ï¼šè§„æ¨¡ä¸å¤§æ—¶å­˜ take è¡¨
    if (n + 1) * (C + 1) <= 2_000_000:
        take = [[False] * (C + 1) for _ in range(n)]
        for i, it in enumerate(items):
            w, v = it.w, it.v
            for c in range(C, w - 1, -1):
                if dp[c - w] + v > dp[c]:
                    dp[c] = dp[c - w] + v
                    take[i][c] = True
        c = max(range(C + 1), key=lambda x: dp[x])
        chosen: List[int] = []
        for i in range(n - 1, -1, -1):
            if take[i][c]:
                chosen.append(items[i].idx)
                c -= items[i].w
        return max(dp), chosen[::-1]
    else:
        # ä¸å›æº¯ç‰ˆæœ¬ï¼ˆä»…å€¼ï¼‰
        for it in items:
            w, v = it.w, it.v
            for c in range(C, w - 1, -1):
                if dp[c - w] + v > dp[c]:
                    dp[c] = dp[c - w] + v
        return max(dp), []

def fractional_upper_bound(items: List[Item], C: int) -> float:
    """LP/åˆ†æ•°èƒŒåŒ…ä¸Šç•Œï¼šæŒ‰ v/w æ’åºï¼Œæœ€åä¸€ä¸ªç‰©å“æŒ‰æ¯”ä¾‹å¡«å……ã€‚"""
    items_sorted = sorted(items, key=lambda x: (x.v / x.w, x.v), reverse=True)
    tw = 0
    tv = 0.0
    for it in items_sorted:
        if tw + it.w <= C:
            tw += it.w
            tv += it.v
        else:
            remain = C - tw
            if remain > 0:
                tv += (it.v / it.w) * remain
            break
    return tv


#  äº”ã€ç»Ÿä¸€çš„"è¿‘å®¹é‡å¯¹æŠ—æ¨¡æ¿"å·¥å…· - ä¸åŸå§‹å®éªŒå®Œå…¨ç›¸åŒ


def _build_near_capacity_items(
    n: int,
    rng: random.Random,
    C: int,
    w_lo: float = 0.55,
    w_hi: float = 0.75
) -> List[int]:
    """ç”Ÿæˆ n ä¸ª'è¿‘å®¹é‡'é‡é‡ï¼Œæ¯ä»¶ > 0.5Cï¼Œä»è€Œä»»ä½•ä¸¤ä»¶æ— æ³•åŒæ—¶è£…å…¥ã€‚"""
    return [rng.randint(int(w_lo * C), int(w_hi * C)) for _ in range(n)]

def _finish_with_big_item(
    items_small: List[Item],
    C: int,
    rng: random.Random,
    density_eps: float = 0.01,
    big_w_ratio: float = 0.95
) -> List[Item]:
    """
    åœ¨å°ä»¶åŸºç¡€ä¸Šæ·»åŠ ä¸€ä¸ª"å¤§ä»¶"ï¼š

    - w_big = big_w_ratio * C ï¼ˆé»˜è®¤ 0.95Cï¼‰
      ç¡®ä¿ä»»æ„è£…å…¥ä¸€ä»¶å°ä»¶åï¼Œå¤§ä»¶æ— æ³•å†è£…å…¥
    - d_big = d_best * (1 - eps)
      å¯†åº¦ç•¥ä½äºæœ€ä½³å°ä»¶ï¼Œæ’åºæ—¶å¤§ä»¶é å
    - v_big = ceil(d_big * w_big)
      ç”±äº w_big >> w_smallï¼Œå³ä½¿ d_big < d_bestï¼Œé€šå¸¸ v_big > v_best_small
    """
    best_small = max(items_small, key=lambda it: it.v / it.w)
    d_best = best_small.v / best_small.w

    w_big = int(max(1, round(big_w_ratio * C)))
    d_big = max(1e-9, d_best * (1.0 - density_eps))
    v_big = int(math.ceil(d_big * w_big))

    items = items_small + [
        Item(w=w_big, v=max(1, v_big), idx=len(items_small))
    ]
    return items


#  å…­ã€å®ä¾‹ç”Ÿæˆå™¨ - ä¸åŸå§‹å®éªŒå®Œå…¨ç›¸åŒ


C_BASE_DEFAULT = 1000

def gen_uncorrelated(
    n: int,
    seed: int,
    cap_ratio: float = 0.5,
    C_base: int = C_BASE_DEFAULT,
    v_lo: int = 1,
    v_hi: int = 1000
):
    rng = random.Random(seed)
    C = int(C_base)
    ws = _build_near_capacity_items(n - 1, rng, C, 0.55, 0.75)
    vs = [rng.randint(v_lo, v_hi) for _ in range(n - 1)]
    items_small = [Item(ws[i], vs[i], i) for i in range(n - 1)]
    items = _finish_with_big_item(items_small, C, rng, density_eps=0.01, big_w_ratio=0.95)
    return items, C

def gen_weakly_correlated(
    n: int,
    seed: int,
    cap_ratio: float = 0.5,
    C_base: int = C_BASE_DEFAULT,
    noise: int = 20
):
    rng = random.Random(seed)
    C = int(C_base)
    ws = _build_near_capacity_items(n - 1, rng, C, 0.55, 0.75)
    vs = [max(1, w + rng.randint(-noise, noise)) for w in ws]
    items_small = [Item(ws[i], vs[i], i) for i in range(n - 1)]
    items = _finish_with_big_item(items_small, C, rng, density_eps=0.01, big_w_ratio=0.95)
    return items, C

def gen_strongly_correlated(
    n: int,
    seed: int,
    cap_ratio: float = 0.5,
    C_base: int = C_BASE_DEFAULT,
    delta: int = 40
):
    rng = random.Random(seed)
    C = int(C_base)
    ws = _build_near_capacity_items(n - 1, rng, C, 0.55, 0.75)
    vs = [w + delta for w in ws]  # å¼ºç›¸å…³ï¼šv = w + Î”
    items_small = [Item(ws[i], vs[i], i) for i in range(n - 1)]
    items = _finish_with_big_item(items_small, C, rng, density_eps=0.01, big_w_ratio=0.95)
    return items, C

def gen_inversely_correlated(
    n: int,
    seed: int,
    cap_ratio: float = 0.5,
    C_base: int = C_BASE_DEFAULT
):
    rng = random.Random(seed)
    C = int(C_base)
    ws = _build_near_capacity_items(n - 1, rng, C, 0.55, 0.75)
    Wmax = max(ws)
    vs = [max(1, Wmax - w + rng.randint(0, 20)) for w in ws]  # åç›¸å…³ï¼šè¶Šè½»è¶Šå€¼é’±
    items_small = [Item(ws[i], vs[i], i) for i in range(n - 1)]
    items = _finish_with_big_item(items_small, C, rng, density_eps=0.01, big_w_ratio=0.95)
    return items, C

def gen_heavy_tailed(
    n: int,
    seed: int,
    cap_ratio: float = 0.5,
    C_base: int = C_BASE_DEFAULT,
    pareto_alpha: float = 2.0
):
    rng = random.Random(seed)
    C = int(C_base)
    ws = []
    for _ in range(n - 1):
        raw = int(C * min(0.95, max(0.55, rng.paretovariate(pareto_alpha) / (2.0 * 5.0))))
        w = int(min(max(0.55 * C, raw), 0.75 * C))
        ws.append(w)
    vs = [rng.randint(200, 1200) for _ in range(n - 1)]
    items_small = [Item(ws[i], vs[i], i) for i in range(n - 1)]
    items = _finish_with_big_item(items_small, C, rng, density_eps=0.015, big_w_ratio=0.95)
    return items, C

def gen_greedy_trap(
    n: int,
    seed: int,
    cap_ratio: float = 0.5,
    C_base: int = C_BASE_DEFAULT
):
    """ç»å…¸é™·é˜±ï¼šå°ä»¶å¯†åº¦ç•¥é«˜ä½†éƒ½å¾ˆé‡ï¼ˆ>0.5Cï¼‰ï¼Œåªèƒ½æ‹¿ 1 ä»¶ï¼›å¤§ä»¶å¯†åº¦ç•¥ä½ä½†æ›´å€¼é’±ã€‚"""
    rng = random.Random(seed)
    C = int(C_base)
    ws = _build_near_capacity_items(n - 1, rng, C, 0.55, 0.75)
    vs = [int(math.ceil(1.02 * w + rng.uniform(-0.01, 0.01) * C)) for w in ws]
    items_small = [Item(ws[i], vs[i], i) for i in range(n - 1)]
    items = _finish_with_big_item(items_small, C, rng, density_eps=0.02, big_w_ratio=0.95)
    return items, C

# --- ä¸¥æ ¼å¯¹æŠ—ï¼ˆGreedyTrap-ADVï¼‰ ---

ADV_DELTA = 0.003
ADV_C_BASE = 1000

def gen_greedy_trap_adv(
    n: int,
    seed: int,
    cap_ratio: float = 0.5,
    delta: float = ADV_DELTA,
    C_base: int = ADV_C_BASE,
    w_lo: float = 0.90,
    w_hi: float = 0.98,
    d_small: float = 1.01,
    noise_frac: float = 0.002
):
    """ä¸¥æ ¼å¯¹æŠ—ï¼šn-1 ä¸ªå°ä»¶ï¼Œæ¯ä»¶ > 0.9Cï¼›1 ä¸ªå¤§ä»¶é‡ä¸º Cï¼Œå¯†åº¦ç•¥ä½ä½†ä»·å€¼æ›´é«˜ã€‚"""
    rng = random.Random(seed)
    C = int(C_base)
    n_small = max(1, n - 1)
    items: List[Item] = []
    for i in range(n_small):
        w = rng.randint(int(w_lo * C), int(w_hi * C))
        noise = rng.uniform(-noise_frac, noise_frac) * C
        v = int(math.ceil(d_small * w + noise))
        items.append(Item(w=max(1, w), v=max(1, v), idx=i))
    d_big = 0.95 * d_small + delta
    v_big = int(math.ceil(d_big * C))
    items.append(Item(w=C, v=max(1, v_big), idx=n_small))
    return items, C

def gen_uncorrelated_large(
    n: int,
    seed: int,
    cap_ratio: float = 0.5,
    C_base: int = C_BASE_DEFAULT
):
    """å¤§è§„æ¨¡ Uncorrelatedï¼šç”¨äºå¯æ‰©å±•æ€§ï¼Œç¦ç”¨ DPã€‚"""
    rng = random.Random(seed)
    C = int(C_base)
    ws = _build_near_capacity_items(n - 1, rng, C, 0.55, 0.75)
    vs = [rng.randint(1, 1000) for _ in range(n - 1)]
    items_small = [Item(ws[i], vs[i], i) for i in range(n - 1)]
    items = _finish_with_big_item(items_small, C, rng, density_eps=0.01, big_w_ratio=0.95)
    return items, C

def gen_sa_test_medium(
    n: int,
    seed: int,
    cap_ratio: float = 0.6,  # ç¨å®½æ¾çš„å®¹é‡
    C_base: int = C_BASE_DEFAULT
):
    """ä¸­ç­‰è§„æ¨¡æµ‹è¯•å®ä¾‹ - å¹³è¡¡æœç´¢éš¾åº¦"""
    rng = random.Random(seed)
    C = int(C_base * cap_ratio)
    
    # ç”Ÿæˆå¤šæ ·åŒ–çš„ç‰©å“
    ws = []
    vs = []
    
    # 30% å°é‡é‡ç‰©å“
    small_count = int(n * 0.3)
    for _ in range(small_count):
        w = rng.randint(1, int(0.2 * C))
        v = rng.randint(1, 100)
        ws.append(w)
        vs.append(v)
    
    # 40% ä¸­ç­‰é‡é‡ç‰©å“  
    medium_count = int(n * 0.4)
    for _ in range(medium_count):
        w = rng.randint(int(0.2 * C), int(0.5 * C))
        v = rng.randint(50, 300)
        ws.append(w)
        vs.append(v)
    
    # 30% å¤§é‡é‡ç‰©å“
    large_count = n - small_count - medium_count
    for _ in range(large_count):
        w = rng.randint(int(0.5 * C), int(0.8 * C))
        v = rng.randint(200, 500)
        ws.append(w)
        vs.append(v)
    
    items = [Item(ws[i], vs[i], i) for i in range(n)]
    return items, C

def gen_sa_test_large(
    n: int,
    seed: int,
    cap_ratio: float = 0.7,
    C_base: int = C_BASE_DEFAULT
):
    """å¤§è§„æ¨¡æµ‹è¯•å®ä¾‹ - æµ‹è¯•å¯æ‰©å±•æ€§"""
    rng = random.Random(seed)
    C = int(C_base * cap_ratio)
    
    ws = []
    vs = []
    
    # æ›´å¤æ‚çš„ä»·å€¼-é‡é‡å…³ç³»
    for i in range(n):
        base_weight = rng.randint(1, int(0.6 * C))
        
        # ä¸‰ç§ç›¸å…³æ¨¡å¼æ··åˆ
        pattern = rng.choice(['uncorrelated', 'weakly_correlated', 'strongly_correlated'])
        
        if pattern == 'uncorrelated':
            value = rng.randint(1, 500)
        elif pattern == 'weakly_correlated':
            value = max(1, base_weight + rng.randint(-50, 50))
        else:  # strongly_correlated
            value = base_weight + rng.randint(30, 100)
            
        ws.append(base_weight)
        vs.append(value)
    
    items = [Item(ws[i], vs[i], i) for i in range(n)]
    return items, C

# æ‰€æœ‰å®ä¾‹æ—çš„ç»Ÿä¸€æ³¨å†Œè¡¨ - ä¸åŸå§‹å®éªŒå®Œå…¨ç›¸åŒ
GENS = {
    "Uncorrelated":       gen_uncorrelated,
    "WeaklyCorr":         gen_weakly_correlated,
    "StronglyCorr":       gen_strongly_correlated,
    "InverseCorr":        gen_inversely_correlated,
    "HeavyTailed":        gen_heavy_tailed,
    "GreedyTrap":         gen_greedy_trap,
    "GreedyTrap-ADV":     gen_greedy_trap_adv,
    "Uncorrelated-Large": gen_uncorrelated_large,
    "SA-Test-Medium":     gen_sa_test_medium,
    "SA-Test-Large":      gen_sa_test_large,
}


#  ä¸ƒã€å•å®ä¾‹è¯„æµ‹å‡½æ•° - ä¸åŸå§‹å®éªŒç›¸åŒ


def run_instance(items: List[Item], C: int, allow_dp: bool = True):
    """åœ¨å•ä¸ªå®ä¾‹ä¸Šè¯„æµ‹æ‰€æœ‰ ALGOS ä¸­æ³¨å†Œçš„ç®—æ³•ï¼Œè¿”å›å¤šè¡Œè®°å½•ã€‚"""
    n = len(items)
    total_w = sum(it.w for it in items)

    # é€‰æ‹©åŸºçº¿ï¼šå°/ä¸­è§„æ¨¡ç”¨ DP å¾— OPTï¼›å¦åˆ™ç”¨ LP ä¸Šç•Œ
    if allow_dp and n <= 200 and C <= 6000:
        t0 = time.perf_counter()
        opt_val, _ = dp_optimal(items, C)
        opt_time = time.perf_counter() - t0
        opt_known = True
        UB = float(opt_val)
    else:
        opt_val = None
        opt_time = None
        opt_known = False
        UB = fractional_upper_bound(items, C)

    rows = []
    for name, algo in ALGOS.items():
        t0 = time.perf_counter()
        val, _ = algo(items, C)
        runtime = time.perf_counter() - t0

        if opt_known and UB > 0:
            gap = (opt_val - val) / opt_val if opt_val > 0 else 0.0
        else:
            gap = (UB - val) / UB if UB > 0 else 0.0

        rows.append({
            "algo":       name,
            "value":      int(val),
            "runtime_s":  float(runtime),
            "gap":        float(gap),
            "opt_known":  bool(opt_known),
            "opt_time_s": (float(opt_time) if opt_time is not None else None),
            "capacity":   int(C),
            "n":          int(n),
            "sum_w":      int(total_w),
        })
    return rows


#  å…«ã€ä¸»å®éªŒå¾ªç¯ - ä¸åŸå§‹å®éªŒç›¸åŒ


def run_benchmark(seed_base: int = 2025) -> pd.DataFrame:
    """
    ä¸»å®éªŒå…¥å£ï¼š
    - å›ºå®š classes & GENS å®šä¹‰æ•°æ®é›†
    - å¯¹æ¯ä¸ª (class, rep, sample) ç”Ÿæˆå®ä¾‹å¹¶è°ƒç”¨ run_instance
    - å°†ç»“æœä¿å­˜ä¸º CSV
    """

    classes = [
        ("Uncorrelated",       dict(n=60)),
        ("WeaklyCorr",         dict(n=150)),
        ("StronglyCorr",       dict(n=150)),
        ("InverseCorr",        dict(n=150)),
        ("HeavyTailed",        dict(n=150)),
        ("GreedyTrap",         dict(n=150)),
        ("GreedyTrap-ADV",     dict(n=13)),   # 12 å°ä»¶ + 1 å¤§ä»¶ï¼Œä¸¥æ ¼å¯¹æŠ—
        ("Uncorrelated-Large", dict(n=500)),  # å¤§è§„æ¨¡ç±»ï¼Œä¸åš DP
        # æ¨¡æ‹Ÿé€€ç«ä¸“ç”¨æµ‹è¯•é›†
        ("SA-Test-Medium",     dict(n=200)),
        ("SA-Test-Large",      dict(n=1000)),
    ]

    REPEATS = 5
    SAMPLES_PER_REPEAT = 20

    rows = []
    for cls_name, params in classes:
        gen = GENS[params.get("gen", cls_name)]
        for rep in range(REPEATS):
            for s in range(SAMPLES_PER_REPEAT):
                # ä½¿ç”¨ç¨³å®šå“ˆå¸Œç”Ÿæˆå­ç§å­ï¼šç¡®ä¿åŒ (class, rep, sample) è·¨ç®—æ³•ä¸€è‡´
                seed = seed_base + (stable_hash_str(cls_name, rep, s) % 10_000_000)
                items, C = gen(params["n"], seed)
                allow_dp = (cls_name != "Uncorrelated-Large" and cls_name != "SA-Test-Large")
                res = run_instance(items, C, allow_dp=allow_dp)
                for r in res:
                    r2 = {"class": cls_name, "rep": rep, "sample": s}
                    r2.update(r)
                    rows.append(r2)

    df = pd.DataFrame(rows)
    out_csv = os.path.join(OUTDIR, "sa_memetic_results.csv")
    df.to_csv(out_csv, index=False)
    print(f"[OK] Results saved to: {out_csv}")
    print(df.head(3))
    return df


#  ä¹ã€å¯¹æ¯”åˆ†æå‡½æ•° - ä¿®æ”¹åçš„ç‰ˆæœ¬

def analyze_comparison(df):
    """åˆ†æSAå’ŒMemetic SAçš„å¯¹æ¯”ç»“æœ"""
    
    print("\n" + "="*60)
    print("æ¨¡æ‹Ÿé€€ç«(SA) vs è®°å¿†æ¨¡æ‹Ÿé€€ç«(Memetic SA) å¯¹æ¯”åˆ†æ")
    print("="*60)
    
    # 1. æ•´ä½“æ€§èƒ½å¯¹æ¯”
    overall_stats = df.groupby('algo').agg({
        'gap': ['mean', 'std', 'min', 'max'],
        'runtime_s': ['mean', 'std', 'min', 'max'],
        'value': 'mean'
    }).round(4)
    
    print("\næ•´ä½“æ€§èƒ½å¯¹æ¯”:")
    print(overall_stats)
    
    # 2. æŒ‰å®ä¾‹ç±»å‹åˆ†æ
    class_comparison = df.groupby(['class', 'algo']).agg({
        'gap': 'mean',
        'runtime_s': 'mean',
        'value': 'mean'
    }).unstack().round(4)
    
    print("\næŒ‰å®ä¾‹ç±»å‹å¯¹æ¯” (å¹³å‡Gap):")
    print(class_comparison['gap'])
    
    print("\næŒ‰å®ä¾‹ç±»å‹å¯¹æ¯” (å¹³å‡è¿è¡Œæ—¶é—´):")
    print(class_comparison['runtime_s'])
    
    # 3. ç»Ÿè®¡æ˜¾è‘—æ€§æ£€éªŒ
    print("\næ€§èƒ½æå‡ç»Ÿè®¡:")
    for cls in df['class'].unique():
        cls_data = df[df['class'] == cls]
        sa_gaps = cls_data[cls_data['algo'] == 'SA_Fast']['gap'].values
        memetic_gaps = cls_data[cls_data['algo'] == 'MemeticSA_Fast']['gap'].values
        
        if len(sa_gaps) > 0 and len(memetic_gaps) > 0:
            improvement = np.mean(sa_gaps) - np.mean(memetic_gaps)
            improvement_pct = (improvement / np.mean(sa_gaps)) * 100 if np.mean(sa_gaps) > 0 else 0
            
            sa_time = cls_data[cls_data['algo'] == 'SA_Fast']['runtime_s'].mean()
            memetic_time = cls_data[cls_data['algo'] == 'MemeticSA_Fast']['runtime_s'].mean()
            time_ratio = memetic_time / sa_time if sa_time > 0 else 1
            
            print(f"{cls}:")
            print(f"  Gapæ”¹è¿›: {improvement:.4f} ({improvement_pct:.1f}%)")
            print(f"  æ—¶é—´æ¯”(Memetic/SA): {time_ratio:.2f}x")
            print(f"  SAå¹³å‡Gap: {np.mean(sa_gaps):.4f}")
            print(f"  Memeticå¹³å‡Gap: {np.mean(memetic_gaps):.4f}")
    
    # 4. å®ä¾‹è§„æ¨¡å¯¹æ€§èƒ½çš„å½±å“
    print("\nå®ä¾‹è§„æ¨¡å¯¹ç®—æ³•æ€§èƒ½çš„å½±å“:")
    df['size_category'] = pd.cut(df['n'], 
                                 bins=[0, 100, 200, 500, 1000, float('inf')],
                                 labels=['å¾®å‹(<100)', 'å°å‹(100-200)', 'ä¸­å‹(200-500)', 
                                         'å¤§å‹(500-1000)', 'è¶…å¤§å‹(>1000)'])
    
    # ä¿®å¤: æ·»åŠ  observed=True å‚æ•°
    size_analysis = df.groupby(['size_category', 'algo'], observed=True).agg({
        'gap': 'mean',
        'runtime_s': 'mean'
    }).unstack().round(4)
    
    print("\nä¸åŒè§„æ¨¡å®ä¾‹ä¸Šçš„å¹³å‡Gap:")
    print(size_analysis['gap'])
    
    print("\nä¸åŒè§„æ¨¡å®ä¾‹ä¸Šçš„å¹³å‡è¿è¡Œæ—¶é—´(s):")
    print(size_analysis['runtime_s'])
    
    return overall_stats, class_comparison, size_analysis

def create_visualizations(df):
    """Create comparative analysis visualization charts"""
    
    # è®¾ç½®ç»˜å›¾æ ·å¼
    plt.style.use('seaborn-v0_8-darkgrid')
    rcParams['font.size'] = 10
    
    # 1. æ•´ä½“Gapå¯¹æ¯”ç®±çº¿å›¾
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # ç®±çº¿å›¾ï¼šæ•´ä½“Gapåˆ†å¸ƒ
    ax = axes[0, 0]
    df_box = df[['algo', 'gap']].copy()
    df_box['gap'] = df_box['gap'].clip(upper=1.0)
    sns.boxplot(x='algo', y='gap', data=df_box, ax=ax)
    ax.set_title('Algorithm Gap Distribution Comparison')
    ax.set_xlabel('Algorithm')
    ax.set_ylabel('Gap Value')
    ax.set_ylim(0, 1.0)
    
    # è¿è¡Œæ—¶é—´å¯¹æ¯”
    ax = axes[0, 1]
    time_data = df[['algo', 'runtime_s']].copy()
    time_data['runtime_s'] = np.log1p(time_data['runtime_s'])
    sns.boxplot(x='algo', y='runtime_s', data=time_data, ax=ax)
    ax.set_title('Algorithm Runtime Comparison (Log Scale)')
    ax.set_xlabel('Algorithm')
    ax.set_ylabel('log(Runtime + 1)')
    
    # å„å®ä¾‹ç±»å‹Gapå¯¹æ¯”
    ax = axes[1, 0]
    gap_by_class = df.groupby(['class', 'algo'])['gap'].mean().unstack()
    gap_by_class.plot(kind='bar', ax=ax, width=0.8)
    ax.set_title('Average Gap Comparison by Instance Type')
    ax.set_xlabel('Instance Type')
    ax.set_ylabel('Average Gap')
    ax.legend(title='Algorithm')
    ax.tick_params(axis='x', rotation=45)
    
    # è§„æ¨¡å¯¹Gapçš„å½±å“
    ax = axes[1, 1]
    df['size_category'] = pd.cut(df['n'], 
                                 bins=[0, 100, 200, 500, 1000, float('inf')],
                                 labels=['<100', '100-200', '200-500', '500-1000', '>1000'])
    size_gap = df.groupby(['size_category', 'algo'], observed=True)['gap'].mean().unstack()
    size_gap.plot(kind='line', marker='o', ax=ax)
    ax.set_title('Impact of Instance Size on Algorithm Gap')
    ax.set_xlabel('Instance Size (Number of Items)')
    ax.set_ylabel('Average Gap')
    ax.legend(title='Algorithm')
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(os.path.join(OUTDIR, 'sa_memetic_comparison.png'), dpi=300, bbox_inches='tight')
    plt.show()
    
    # 2. è¯¦ç»†å¯¹æ¯”çƒ­åŠ›å›¾
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # å‡†å¤‡çƒ­åŠ›å›¾æ•°æ®
    heatmap_data = df.pivot_table(
        index='class',
        columns='algo',
        values='gap',
        aggfunc='mean'
    ).fillna(0)
    
    # åˆ›å»ºçƒ­åŠ›å›¾
    sns.heatmap(heatmap_data, 
                annot=True, 
                fmt='.4f',
                cmap='RdYlGn_r',
                center=0.5,
                ax=ax)
    
    ax.set_title('Algorithm Performance Heatmap by Instance Type (Average Gap)')
    ax.set_xlabel('Algorithm')
    ax.set_ylabel('Instance Type')
    
    plt.tight_layout()
    plt.savefig(os.path.join(OUTDIR, 'performance_heatmap.png'), dpi=300, bbox_inches='tight')
    plt.show()
    
    # 3. è¿è¡Œæ—¶é—´ä¸Gapçš„æƒè¡¡å›¾
    fig, ax = plt.subplots(figsize=(10, 6))
    
    for algo in df['algo'].unique():
        algo_data = df[df['algo'] == algo]
        
        # æŒ‰å®ä¾‹ç±»å‹èšåˆ
        agg_data = algo_data.groupby('class').agg({
            'runtime_s': 'mean',
            'gap': 'mean'
        }).reset_index()
        
        ax.scatter(agg_data['runtime_s'], 
                  agg_data['gap'], 
                  label=algo, 
                  s=100, 
                  alpha=0.7)
        
        # æ·»åŠ æ ‡ç­¾
        for _, row in agg_data.iterrows():
            ax.annotate(row['class'], 
                       (row['runtime_s'], row['gap']),
                       fontsize=8,
                       alpha=0.7)
    
    ax.set_xlabel('Average Runtime (seconds)')
    ax.set_ylabel('Average Gap')
    ax.set_title('Algorithm Performance Trade-off: Runtime vs Gap')
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(os.path.join(OUTDIR, 'time_gap_tradeoff.png'), dpi=300, bbox_inches='tight')
    plt.show()
    
    print(f"[OK] Visualization charts saved to {OUTDIR} directory")

def convert_keys_for_json(obj):
    """å°†å­—å…¸ä¸­çš„å…ƒç»„é”®è½¬æ¢ä¸ºå­—ç¬¦ä¸²ï¼Œä»¥ä¾¿JSONåºåˆ—åŒ–"""
    if isinstance(obj, dict):
        new_dict = {}
        for key, value in obj.items():
            if isinstance(key, tuple):
                # å°†å…ƒç»„è½¬æ¢ä¸ºå­—ç¬¦ä¸²ï¼Œä¾‹å¦‚ï¼š('gap', 'mean') -> 'gap_mean'
                new_key = '_'.join(str(k) for k in key)
            else:
                new_key = key
            new_dict[new_key] = convert_keys_for_json(value)
        return new_dict
    elif isinstance(obj, list):
        return [convert_keys_for_json(item) for item in obj]
    elif isinstance(obj, pd.DataFrame):
        # å¦‚æœæ˜¯DataFrameï¼Œå…ˆè½¬æ¢ä¸ºå­—å…¸
        return convert_keys_for_json(obj.to_dict())
    else:
        return obj

def generate_comprehensive_report(df, overall_stats, class_comparison, size_analysis):
    """ç”Ÿæˆç»¼åˆå¯¹æ¯”æŠ¥å‘Š"""
    
    # ä¿®å¤ï¼šå°†DataFrameè½¬æ¢ä¸ºJSONå¯åºåˆ—åŒ–çš„æ ¼å¼
    overall_stats_dict = convert_keys_for_json(overall_stats)
    size_analysis_dict = convert_keys_for_json(size_analysis)
    
    # è®¡ç®—å…³é”®ç»Ÿè®¡
    best_gap_algo = overall_stats[('gap', 'mean')].idxmin()
    best_gap_value = float(overall_stats[('gap', 'mean')].min())
    fastest_algo = overall_stats[('runtime_s', 'mean')].idxmin()
    fastest_time = float(overall_stats[('runtime_s', 'mean')].min())
    
    report = {
        "å®éªŒä¿¡æ¯": {
            "å®éªŒåç§°": "æ¨¡æ‹Ÿé€€ç«(SA) vs è®°å¿†æ¨¡æ‹Ÿé€€ç«(Memetic SA)å¯¹æ¯”å®éªŒ",
            "å®éªŒæ—¶é—´": time.strftime("%Y-%m-%d %H:%M:%S"),
            "éšæœºç§å­": 42,
            "ç®—æ³•æ•°é‡": len(df['algo'].unique()),
            "å®ä¾‹ç±»å‹æ•°é‡": len(df['class'].unique()),
            "æ€»æµ‹è¯•å®ä¾‹æ•°": len(df) // len(df['algo'].unique()),
            "è¾“å‡ºç›®å½•": OUTDIR
        },
        "ç®—æ³•é…ç½®": {
            "SA_Fast": "æ ‡å‡†æ¨¡æ‹Ÿé€€ç«ç®—æ³•ï¼ˆå¿«é€Ÿç‰ˆï¼‰",
            "MemeticSA_Fast": "è®°å¿†æ¨¡æ‹Ÿé€€ç«ç®—æ³•ï¼ˆå¿«é€Ÿç‰ˆï¼Œæ¨¡æ‹Ÿé€€ç«+å±€éƒ¨æœç´¢+ç§ç¾¤è¿›åŒ–ï¼‰"
        },
        "æ•´ä½“æ€§èƒ½æ‘˜è¦": overall_stats_dict,
        "å…³é”®å‘ç°": {
            "æœ€ä½³å¹³å‡Gapç®—æ³•": best_gap_algo,
            "æœ€ä½³å¹³å‡Gapå€¼": best_gap_value,
            "æœ€å¿«ç®—æ³•": fastest_algo,
            "æœ€å¿«ç®—æ³•å¹³å‡æ—¶é—´": fastest_time,
            "æœ€æ…¢ç®—æ³•": overall_stats[('runtime_s', 'mean')].idxmax(),
            "æœ€æ…¢ç®—æ³•å¹³å‡æ—¶é—´": float(overall_stats[('runtime_s', 'mean')].max())
        },
        "å®ä¾‹ç±»å‹åˆ†æ": {},
        "è§„æ¨¡å½±å“åˆ†æ": size_analysis_dict
    }
    
    # æ·»åŠ å„å®ä¾‹ç±»å‹çš„è¯¦ç»†åˆ†æ
    for cls in df['class'].unique():
        cls_data = df[df['class'] == cls]
        
        # è®¡ç®—æ€§èƒ½æ”¹è¿›
        sa_gaps = cls_data[cls_data['algo'] == 'SA_Fast']['gap'].mean()
        memetic_gaps = cls_data[cls_data['algo'] == 'MemeticSA_Fast']['gap'].mean()
        improvement = sa_gaps - memetic_gaps
        improvement_pct = (improvement / sa_gaps) * 100 if sa_gaps > 0 else 0
        
        report["å®ä¾‹ç±»å‹åˆ†æ"][cls] = {
            "SAå¹³å‡Gap": float(sa_gaps),
            "MemeticSAå¹³å‡Gap": float(memetic_gaps),
            "Gapæ”¹è¿›": float(improvement),
            "Gapæ”¹è¿›ç™¾åˆ†æ¯”": float(improvement_pct),
            "æœ€ä¼˜ç®—æ³•": "MemeticSA_Fast" if memetic_gaps < sa_gaps else "SA_Fast"
        }
    
    # ä¿å­˜æŠ¥å‘Š
    report_path = os.path.join(OUTDIR, "comparison_report.json")
    with open(report_path, 'w', encoding='utf-8') as f:
        json.dump(report, f, ensure_ascii=False, indent=2)
    
    print(f"[OK] ç»¼åˆå¯¹æ¯”æŠ¥å‘Šå·²ä¿å­˜è‡³: {report_path}")
    
    # åŒæ—¶ä¿å­˜æ–‡æœ¬æ ¼å¼çš„æŠ¥å‘Š
    text_report_path = os.path.join(OUTDIR, "comparison_report.txt")
    with open(text_report_path, 'w', encoding='utf-8') as f:
        f.write("="*80 + "\n")
        f.write("æ¨¡æ‹Ÿé€€ç«(SA) vs è®°å¿†æ¨¡æ‹Ÿé€€ç«(Memetic SA) å¯¹æ¯”å®éªŒæŠ¥å‘Š\n")
        f.write("="*80 + "\n\n")
        
        f.write("1. å®éªŒæ¦‚è§ˆ\n")
        f.write("-"*40 + "\n")
        for key, value in report["å®éªŒä¿¡æ¯"].items():
            f.write(f"  {key}: {value}\n")
        
        f.write("\n2. æ•´ä½“æ€§èƒ½å¯¹æ¯”\n")
        f.write("-"*40 + "\n")
        
        # ä»¥æ›´æ˜“è¯»çš„æ ¼å¼æ˜¾ç¤ºæ•´ä½“æ€§èƒ½
        for algo in overall_stats.index:
            f.write(f"\n  {algo}:\n")
            f.write(f"    å¹³å‡Gap: {overall_stats.loc[algo, ('gap', 'mean')]:.4f}\n")
            f.write(f"    å¹³å‡è¿è¡Œæ—¶é—´: {overall_stats.loc[algo, ('runtime_s', 'mean')]:.4f}s\n")
            f.write(f"    å¹³å‡ä»·å€¼: {overall_stats.loc[algo, ('value', 'mean')]:.1f}\n")
        
        f.write("\n3. å…³é”®å‘ç°\n")
        f.write("-"*40 + "\n")
        for key, value in report["å…³é”®å‘ç°"].items():
            f.write(f"  {key}: {value}\n")
        
        f.write("\n4. å„å®ä¾‹ç±»å‹æ€§èƒ½å¯¹æ¯”\n")
        f.write("-"*40 + "\n")
        for cls, stats in report["å®ä¾‹ç±»å‹åˆ†æ"].items():
            f.write(f"\n  {cls}:\n")
            for stat_key, stat_value in stats.items():
                if isinstance(stat_value, float):
                    f.write(f"    {stat_key}: {stat_value:.4f}\n")
                else:
                    f.write(f"    {stat_key}: {stat_value}\n")
    
    print(f"[OK] æ–‡æœ¬æ ¼å¼æŠ¥å‘Šå·²ä¿å­˜è‡³: {text_report_path}")
    
    return report


#  åã€ä¸»ç¨‹åºå…¥å£ - ä¿®å¤åçš„ç‰ˆæœ¬


if __name__ == "__main__":
    print("å¼€å§‹è¿è¡Œæ¨¡æ‹Ÿé€€ç«(SA) vs è®°å¿†æ¨¡æ‹Ÿé€€ç«(Memetic SA)å¯¹æ¯”å®éªŒ...")
    print("="*80)
    
    # è¿è¡ŒåŸºå‡†æµ‹è¯•
    start_time = time.time()
    df = run_benchmark()
    benchmark_time = time.time() - start_time
    
    print(f"\nåŸºå‡†æµ‹è¯•å®Œæˆï¼Œè€—æ—¶: {benchmark_time:.2f}ç§’")
    print(f"æµ‹è¯•å®ä¾‹æ€»æ•°: {len(df)}")
    print(f"ç®—æ³•æ•°é‡: {len(df['algo'].unique())}")
    
    # è¿›è¡Œå¯¹æ¯”åˆ†æ
    print("\nå¼€å§‹åˆ†æå¯¹æ¯”ç»“æœ...")
    overall_stats, class_comparison, size_analysis = analyze_comparison(df)
    
    # åˆ›å»ºå¯è§†åŒ–å›¾è¡¨
    print("\nç”Ÿæˆå¯è§†åŒ–å›¾è¡¨...")
    create_visualizations(df)
    
    # ç”Ÿæˆç»¼åˆæŠ¥å‘Š
    print("\nç”Ÿæˆç»¼åˆå¯¹æ¯”æŠ¥å‘Š...")
    report = generate_comprehensive_report(df, overall_stats, class_comparison, size_analysis)
    
    print("\n" + "="*80)
    print("å®éªŒå®Œæˆï¼")
    print(f"æ‰€æœ‰ç»“æœå·²ä¿å­˜è‡³: {OUTDIR}")
    print("="*80)
    
    # æ˜¾ç¤ºå…³é”®ç»“æœæ‘˜è¦
    print("\nå…³é”®ç»“æœæ‘˜è¦:")
    print("-"*40)
    
    best_gap_algo = overall_stats[('gap', 'mean')].idxmin()
    best_gap_value = overall_stats[('gap', 'mean')].min()
    fastest_algo = overall_stats[('runtime_s', 'mean')].idxmin()
    fastest_time = overall_stats[('runtime_s', 'mean')].min()
    
    print(f"æœ€ä½³Gapç®—æ³•: {best_gap_algo} (å¹³å‡Gap: {best_gap_value:.4f})")
    print(f"æœ€å¿«ç®—æ³•: {fastest_algo} (å¹³å‡æ—¶é—´: {fastest_time:.4f}ç§’)")
    
    # æ£€æŸ¥Memetic SAæ˜¯å¦åœ¨å¤§å¤šæ•°å®ä¾‹ç±»å‹ä¸Šè¡¨ç°æ›´å¥½
    memetic_wins = 0
    total_classes = 0
    
    for cls in df['class'].unique():
        cls_data = df[df['class'] == cls]
        sa_gap = cls_data[cls_data['algo'] == 'SA_Fast']['gap'].mean()
        memetic_gap = cls_data[cls_data['algo'] == 'MemeticSA_Fast']['gap'].mean()
        
        if memetic_gap < sa_gap:
            memetic_wins += 1
        total_classes += 1
    
    print(f"Memetic SAåœ¨ {memetic_wins}/{total_classes} ä¸ªå®ä¾‹ç±»å‹ä¸Šè¡¨ç°æ›´å¥½")
    
    if memetic_wins > total_classes / 2:
        print("ç»“è®º: Memetic SAæ•´ä½“ä¸Šä¼˜äºæ ‡å‡†æ¨¡æ‹Ÿé€€ç«ç®—æ³•")
    else:
        print("ç»“è®º: æ ‡å‡†æ¨¡æ‹Ÿé€€ç«ç®—æ³•åœ¨æŸäº›åœºæ™¯ä¸‹ä»æœ‰ä¼˜åŠ¿")
````
